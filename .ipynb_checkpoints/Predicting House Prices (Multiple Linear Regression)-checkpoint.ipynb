```python
# Import necessary libraries
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
# ------------------------------------------------------
# 1. Load the Dataset
# ------------------------------------------------------
data = pd.read_csv("./dataset/Housing.csv") # Load housing data
df = pd.DataFrame(data) # Convert to DataFrame
df.head()
```

```python
# ------------------------------------------------------
# 2. Prepare Features and Target
# ------------------------------------------------------
# Independent variables (features)
x = df[['area', 'bedrooms', 'bathrooms']]

# Dependent variable (target)
y = df[['price']]
```

```python
# ------------------------------------------------------
# 3. Train-Test Split
# ------------------------------------------------------
# Split the data into training (20%) and testing (80%)
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.2, random_state=0)
```

```python
# visualize data 
x.hist() # feature
plt.show()

y.hist() # source
plt.show()
```

```python
# train model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# make predictions
y_pred = model.predict(x_test)

# evaluate the model
print(f"mean absolute error : ", mean_absolute_error(y_test, y_pred))
print(f"r-squared error : ", r2_score(y_test, y_pred))
```

```python
# predict for a new house
# Area=2500 sqft, 4 bedrooms, 3 bathrooms
new_house = pd.DataFrame([[2500, 4, 3]], columns=['area', 'bedrooms', 'bathrooms'])
predicted_price = model.predict(new_house)
print(f"Predicted price: ${predicted_price.flatten()[0]:.2f}")
```
